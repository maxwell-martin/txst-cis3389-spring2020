```python
# Assignment 3 - Web Scraping - Maxwell Martin and Anam Nawab

# Relevant modules for scraping.
from bs4 import BeautifulSoup
import pandas as pd
import requests

# Url of page to scrape.
url = "https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2334524.m570.l1313.TR12.TRC2.A0.H0.Xmouse.TRS0&_nkw=mouse&_sacat=0&LH_TitleDesc=0&_osacat=0&_odkw=laptop"

# Make HTTP request for page.
page = requests.get(url)

# Confirm we get a valid HTTP response.
print(page)
```

```python
# Convert page HTML to a BeautifulSoup object.
soup = BeautifulSoup(page.content, 'html.parser')

# Confirm soup object contains page's HTML.
print(soup)
```

```python
# Get all product names.
product_names_h3s = soup.find_all(class_="s-item__title")

# Get all the product conditions.
product_conditions_spans = soup.find_all(class_="SECONDARY_INFO")

# Get all product prices.
product_prices_spans = soup.find_all(class_="s-item__price")

# Confirm we got the same number of items for each list and the number matches the number of items shown on the page.
print(len(product_names_h3s))
print(len(product_conditions_spans))
print(len(product_prices_spans))
```

```python
# Create list of product names.
product_names = []
for h3 in product_names_h3s:
    print(h3.text)
    product_names.append(h3.text.strip())
    
# Confirm list has all product conditions.
print(product_names)
```

```python
# Create list of product conditions.
product_conditions = []
for span in product_conditions_spans:
    print(span.text)
    product_conditions.append(span.text.strip())
    
# Confirm list contains all product conditions.
print(product_conditions)
```

```python
# Create list of product prices.
product_prices = []
for span in product_prices_spans:
    print(span.text)
    product_prices.append(span.text.strip())
    
# Confirm list contains all product conditions.
print(product_prices)
```

```python
# Create Pandas data frame.
products = pd.DataFrame( { "Product Name": product_names, 
                           "Product Condition": product_conditions, 
                           "Product Price": product_prices } )

# Confirm data frame is setup properly.
print(products)

# Export data frame to a csv file.
csv = products.to_csv("ebay_products.csv", index = None, header = True)
```

```python
df = df.dropna() # Drop the rows with missing values.
```

```python
df.shape # There are 183 rows left in the data frame after dropping rows with missing values.
```

```python
# e.
df['Survived'] = df['Survived'].map({0: 'no', 1: 'yes'})
df
```

```python
from matplotlib import pyplot as plt

# f.
plt.bar(df['Gender'], df['Age'], color = 'y')
plt.title('Age compared by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')

# Male age bar is higher.
```

```python
# g.
plt.bar(df['Survived'],df['Age'], color = 'g')
plt.title('Age compared by Survived')
plt.xlabel('Survived')
plt.ylabel('Age')

# The chart makes it appear that higher aged people suvived more than lower aged people.
# Although, the difference in age between people who survived and did not survive is not that large.
# I do not think this plot gives enough evidence to say that older people were more likely to survive than younger people.
```

```python
# h.
Family_members = df['SibSp'] + df['Parch']
df['Family_members'] = Family_members
df
```

```python
# i.
df = df.drop(['PassengerId', 'Cabin', 'SibSp', 'Parch'], axis = 1)
df
```

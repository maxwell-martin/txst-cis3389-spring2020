```python
# d)
frame1['Status'] = ['yes','yes','yes','yes','no','yes'] # Add to end of table manually, last value should be yes, not no.
frame1
```

```python
frame1 = frame1.drop('Status', axis = 1) # Drop column so we can add it another way
frame1
```

```python
frame1.insert(2, "Status", ['yes','yes','yes','yes','no','yes'], True) # Add column to second postion
frame1
```

```python
#frame1['Status'] = [True, True, True, True, False, True] # Manually update data in Status column
frame1['Status'] = frame1['Status'].map({'yes': True, 'no': False}) # Using the map function to update Status column
frame1
```

```python
# e)
frame1.sort_values(by=['Salary'], ascending = True)
```

```python
frame1.sort_values(by=['Dept', 'Salary'], ascending=[False, False]) # Sort by Dept, then by Salary; both descending.
```

```python
frame1_clean = frame1.drop_duplicates() # Remove duplicate Charlie record
frame1_clean
```

```python
# f)
frame1_clean['Interest'] = frame1_clean['Salary'] * 0.05
frame1_clean
```

```python
# g)
frame2 = frame1_clean.loc[:,['Dept', 'Salary', 'Interest']] # Locate all rows with specific columns and put in new data frame.
frame2
```

```python
# h)
grouped = frame2['Interest'].groupby(frame2['Dept']) # Group Interest column values by Dept.
grouped.mean() # Get mean Interest for each Dept.
```

```python
# i)
frame2[['Salary','Interest']].agg(['mean','std']) # Get mean and std for both columns at the same time.
```

```python
# Practice Problem 1

import pandas as pd
from matplotlib import pyplot as plt
```

```python
# a)
df = pd.DataFrame({"Year": [2000, 2001, 2002, 2003, 2004, 2005],
                   "AccountingMinSalary": [85000, 75000, 50000, 70000, 80000, 70500],
                   "AccountingMaxSalary": [150000, 130000, 80000, 120000, 130000, 123000],
                   "EngineeringMinSalary": [75000, 705000, 47000, 65000, 81000, 70000],
                   "EngineeringMaxSalary": [140000, 120000, 70000, 110000, 130100, 122000]}, 
                   index = ["a", "b", "c", "d", "e", "f"])
df
```

```python
# c)
plt.bar(df['Year'],df['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(df['Year'],df['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.title('Accounting vs Engineering Maximum Salary')
plt.xlabel('Year')
plt.ylabel('Maximum Salary')
plt.legend(loc='best') # best is also 0
```

```python
# d)
plt.bar(df['Year'],df['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.bar(df['Year'],df['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Accounting vs Engineering Minimum Salary')
plt.xlabel('Year')
plt.ylabel('Minimum Salary')
plt.legend(loc='best')
```

```python
# e)
plt.bar(df['Year'],df['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(df['Year'],df['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.title('Accounting Max vs Min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')
```

```python
# f)
plt.bar(df['Year'],df['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.bar(df['Year'],df['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Engineering Max vs Min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')
```

```python
# g)
plt.pie(df['AccountingMaxSalary'], labels=df['Year'],colors=('r','y','m','b'))
plt.title('Pie Chart for Accounting Maximum Salary')
plt.legend(loc='best')
```

```python
# Arrange all plots in one figure
plt.figure(figsize=(20,25)) # Set the size of overall figure

plt.subplot(3,2,1) # Create subplot. This determines the position of the chart in the figure.
plt.bar(df['Year'],df['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(df['Year'],df['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.title('Accounting vs Engineering Maximum Salary')
plt.xlabel('Year')
plt.ylabel('Maximum Salary')
plt.legend(loc='best')

plt.subplot(3,2,2)
plt.bar(df['Year'],df['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.bar(df['Year'],df['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Accounting vs Engineering Minimum Salary')
plt.xlabel('Year')
plt.ylabel('Minimum Salary')
plt.legend(loc='best')

plt.subplot(3,2,3)
plt.bar(df['Year'],df['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(df['Year'],df['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.title('Accounting Max vs Min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')

plt.subplot(3,2,4)
plt.bar(df['Year'],df['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.bar(df['Year'],df['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Engineering Max vs Min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')

plt.subplot(3,2,5)
plt.pie(df['AccountingMaxSalary'], labels=df['Year'],colors=('r','y','m','b'))
plt.title('Pie Chart for Accounting Maximum Salary')
plt.legend(loc='best')

plt.savefig('ex1.png') # Can also do .jpeg, .pdf, .jpg, etc.
```

```python
# Trends - How to create line charts
# Arrange all plots in one figure
plt.figure(figsize=(20,25)) # Set the size of overall figure

plt.subplot(3,2,1) # Create subplot. This determines the position of the chart in the figure.
plt.plot(df['Year'],df['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.plot(df['Year'],df['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.title('Accounting vs Engineering Maximum Salary')
plt.xlabel('Year')
plt.ylabel('Maximum Salary')
plt.legend(loc='best')

plt.subplot(3,2,2)
plt.plot(df['Year'],df['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.plot(df['Year'],df['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Accounting vs Engineering Minimum Salary')
plt.xlabel('Year')
plt.ylabel('Minimum Salary')
plt.legend(loc='best')
```

```python
# Practice Problem 2

df = pd.read_csv("athletes.csv")
df.head(3)
```

```python
plt.figure(figsize=(30,25)) # Set the size of overall figure

plt.subplot(2,2,1)
plt.bar(df['gender'],df['gold'], label='Gold')
plt.bar(df['gender'],df['silver'], label='Silver')
plt.title('Medals by Gender')
plt.xlabel('Gender')
plt.ylabel('Medals')
plt.legend(loc='best')
# Gold is higher for male than female, but silver is the same for both male and female

plt.subplot(2,2,2)
plt.bar(df['gender'],df['age'])
plt.title('Ahtlete age')
plt.xlabel('Gender')
plt.ylabel('Age')
# There is not a significant difference in age between male and female.

plt.subplot(2,2,3)
plt.bar(df['sport'],df['gold'], label='Gold')
plt.bar(df['sport'],df['silver'], label='Silver')
plt.title('Medals per Sport')
plt.xlabel('Sport')
plt.ylabel('Medals')
plt.legend(loc='best')
# There are more sports with silver medals than sports with gold medals.
```
